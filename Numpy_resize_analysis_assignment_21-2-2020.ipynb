# Change its length and height using define (def) without change his real image  shape

```python
def plti (im , h=8 ,**kwargs):
    """""helper function to plot his image."""
    y = im.shape[0]
    x = im.shape[1]
    w = (y/x) * h
    plt.figure(figsize=(w,h))
    plt.imshow(im , interpolation="none" ,**kwargs)
    plt.axis("off")
    input("Press <Enter> to show image ")

    
```
